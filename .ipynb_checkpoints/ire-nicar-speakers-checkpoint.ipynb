```python
#IRE
#2018
from lxml import html
import requests

import pandas as pd
```

```python
page = requests.get('https://www.ire.org/conferences/ire-2018/speakers/')
tree = html.fromstring(page.content)
```

```python
#2018 speakers:
speakers = tree.xpath('//h3[@class="title3"]/text()')
```

```python
speakers
```
